```python
import pandas as pd

# Load the dataset
# Note: Ensure amazon.csv is in the same folder as your Jupyter Notebook
df = pd.read_csv('amazon.csv')

# Display the first 5 rows to see what the data looks like
df.head()
```

```python
# Check the number of rows and columns
print("Dataset Shape:", df.shape)

# Check columns, non-null counts, and current data types
df.info()
```

```python
# Clean column names: convert to lowercase and replace spaces with underscores
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Verify the updated column names
print("Updated Columns:", df.columns.tolist())
```

```python
import pandas as pd

# 1. Improved cleaning function using pd.to_numeric for safety
def clean_price(price_str):
    if pd.isna(price_str):
        return price_str
    # Remove '₹' and ',' characters, and strip whitespaces from individual strings
    return str(price_str).replace('₹', '').replace(',', '').strip()

# Apply price cleaning to the columns
df['discounted_price'] = df['discounted_price'].apply(clean_price)
df['actual_price'] = df['actual_price'].apply(clean_price)

# Convert to float safely (errors='coerce' turns non-numeric mistakes into NaN instead of crashing)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')


# 2. FIXED Line: Notice the added '.str' before '.strip()'
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%', '').str.strip()

# Convert discount percentage to float safely
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')


# 3. Verify your changes
print(df[['discounted_price', 'actual_price', 'discount_percentage']].dtypes)
```

```python
# Check unique values in rating to find anomalies
print("Unique ratings before cleaning:", df['rating'].unique())

# If there is a non-numeric character (e.g., '|' or a blank string), replace it with NaN or a default value
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Check rating_count: remove commas and convert to float/int
df['rating_count'] = df['rating_count'].astype(str).str.replace(',', '').strip()
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

print("Data types after cleaning numbers:\n", df[['rating', 'rating_count']].dtypes)
```

```python
# Identify missing values per column
print("Missing values per column before handling:")
print(df.isnull().sum())

# Strategy: Fill missing ratings or rating_counts with their median values
if df['rating'].isnull().sum() > 0:
    df['rating'].fillna(df['rating'].median(), inplace=True)

if df['rating_count'].isnull().sum() > 0:
    df['rating_count'].fillna(df['rating_count'].median(), inplace=True)

# Double check to ensure no more missing values exist in those critical columns
print("\nMissing values after handling:")
print(df[['rating', 'rating_count']].isnull().sum())
```

```python
# Check for duplicate rows based on unique identifier or entire rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows found: {duplicate_count}")

# Remove duplicates (keeping the first occurrence)
df.drop_duplicates(inplace=True)
print("Duplicates removed successfully.")
```

```python
# Trim spaces from text columns
text_cols = ['product_id', 'product_name', 'category', 'user_id', 'user_name']
for col in text_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()

# Optional: Split the main category from sub-categories if needed for cleaner analysis
# E.g., "Computers&Accessories|Accessories..." -> "Computers&Accessories"
df['main_category'] = df['category'].apply(lambda x: x.split('|')[0] if '|' in x else x)

df[['category', 'main_category']].head()
```

```python
# Save to a new clean CSV file without saving the index column
df.to_csv('amazon_cleaned.csv', index=False)
print("Cleaned dataset saved as 'amazon_cleaned.csv'!")
```
